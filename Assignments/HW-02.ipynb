### **What is an A/B test?** 
A/B tests are experiments that compare two scenarios (e.g., scenario A vs. scenario B). 
They test subjects' responses to each of the variants to determine which is more effective ([read more about A/B tests on Wikipedia](https://en.wikipedia.org/wiki/A/B_testing)).  
A/B tests allow us to draw conclusions about the different scenarios by randomizing exposure to them, e.g., one could flip a coin and assign a user to scenario A if it lands heads and to B if it lands tails.  
Since exposure is randomized, we can be confident that the scenarios are the sole explanation for statistically significant differences in subjects' responses (if they exist).  
In theory, A/B testing refers to an experiment that compares two scenarios; however, in practice, the term is also used when we compare multiple scenarios (e.g., A vs. B vs. C), although the more precise terminology would be to call such an experiment a "multinomial test."

### **How were A/B tests used by Upworthy?** 
Upworthy used A/B testing to increase news readership, conducting experiments for each published news piece.  
In each experiment, they created multiple "packages" of stimuli, varying headlines, images, excerpts, and **ledes** for the same news piece.  
Different "packages" were shown on their (now defunct) website to engage users with the news pieces they produced. Upworthy found "the best" package by conducting A/B tests, showing different packages to different users, and measuring how often users clicked on each version.  
Below, we show three "packages" used by Upworthy in an experiment, each with a different headline for the same news piece.  
Upworthy randomized users that visited their website saw one of the three versions of the headline below. Then, they measured the percentage of times users in each scenario clicked to read the news.  
The headline with the highest percentage of clicks per view (click through rate) was then declared the "winner" and became the default for all visitors.

  1. how many users were shown each package.
  2. how many individuals clicked each package.

1.7 For our subsequent analysis, we want to compare the causal effect of headlines on the success of a news piece. 
For that, we can compare pairs of packages with the same `eyecatcher_id`, `lede`, and `excerpt`, but different `headlines`.
Note that this means that if an experiment considered 5 different headlines and did not vary any other stimulus, we would have 5C2 (i.e., 5 choose 2, 10) pairs to consider.
Create a dataset where:
- Each row corresponds to a pair of packages with different `headline` but the same `eyecatcher_id`, `lede`, and `excerpt`. 
- There are columns containing the headlines of each of the news versions (`headline1`, `headline2`) and the clickthrough rate of each of the news versions (`ctr1`, `ctr2`). 
- The columns `headline1` and `ctr1` contain the data associated with the news version with the highest clickthrough rate. Print the first rows of your newly created dataframe, as well as its length.
-  The rows where the two news pieces had exactly the same clickthrough rate should be filtered out (this is for simplicity's sake).
-  The column `date_created` contains the date when the news version with the highest clickthrough rate was created.


```python
from itertools import combinations

# Filter experiments with more than one headline
multi_headline_experiments = df[df.groupby('clickability_test_id')['headline'].transform('nunique') > 1]
# Initialize an empty list to store the data for the new dataframe
data = []

# Iterate over unique experiments
for _, group in multi_headline_experiments.groupby('clickability_test_id'):
    # Generate pairs of headlines within each experiment
    headline_combinations = combinations(group['headline'].unique(), 2)
    
    # Iterate over pairs and extract data
    for headline1, headline2 in headline_combinations:
        ctr1 = group[group['headline'] == headline1]['ctr'].iloc[0]
        ctr2 = group[group['headline'] == headline2]['ctr'].iloc[0]
        
        # Determine which headline has the higher clickthrough rate
        if ctr1 > ctr2:
            pair_data = {
                'eyecatcher_id': group['eyecatcher_id'].iloc[0],
                'lede': group['lede'].iloc[0],
                'excerpt': group['excerpt'].iloc[0],
                'headline1': headline1,
                'headline2': headline2,
                'ctr1': ctr1,
                'ctr2': ctr2,
                'date_created': group[group['headline'] == headline1]['created_at'].iloc[0]
            }
        else:
            pair_data = {
                'eyecatcher_id': group['eyecatcher_id'].iloc[0],
                'lede': group['lede'].iloc[0],
                'excerpt': group['excerpt'].iloc[0],
                'headline1': headline2,
                'headline2': headline1,
                'ctr1': ctr2,
                'ctr2': ctr1,
                'date_created': group[group['headline'] == headline2]['created_at'].iloc[0]
            }
        
        # Append data to the list
        data.append(pair_data)

# Create the new dataframe
pair_df = pd.DataFrame(data)

# Filter out rows where the two news pieces had the same clickthrough rate
pair_df = pair_df[pair_df['ctr1'] != pair_df['ctr2']]

# Display the first rows of the new dataframe and its length
print(pair_df.head())
print(f"Length of the new dataframe: {len(pair_df)}")
```

```python
# Calculate the average difference per pair between ctr1 and ctr2
average_difference = (pair_df['ctr1'] - pair_df['ctr2']).mean()

# Calculate the average clickthrough rate for the least clicked-through headline (ctr2)
average_ctr2 = pair_df['ctr2'].mean()

# Display the results
print(f"Average difference per pair between ctr1 and ctr2: {average_difference:.4f}")
print(f"Average clickthrough rate for the least clicked-through headline (ctr2): {average_ctr2:.4f}")
```

Each indicator variable in the dataframe should equal 1 if the corresponding headline uses the corresponding type of pronoun and 0 otherwise.

```python
import re

# Dictionary of pronouns
feature_wordsets = dict([
    ('first_person_singular', ['i', 'me', 'my', 'mine', 'myself',
                               "i'd", "i'll", "i'm", "i've", 'id', 'im', 'ive']),
    ('first_person_plural', ['we', 'us', 'our', 'ours', 'ourselves',
                              "we'd", "we'll", "we're", "we've",]),
    ('second_person', ['you', 'your', 'yours', 'yourself',
                       "ya", "you'd", "you'll", "you're", "you've", 'youll', 'youre', 'youve',
                       'yourselves']),
    ('third_person_singular', ['he', 'him', 'his', 'himself',
                               "he'd", "he's", 'hes',
                               'she', 'her', 'hers', 'herself',
                               "she'll", "she's", 'shes',
                               'it', 'its', 'itself',
                               'themself']),
    ('third_person_plural', ['they', 'them', 'their', 'theirs', 'themselves',
                              "they'd", "they'll", "they've", 'theyll', 'theyve'])
])

# Function to create indicator variables for each set of pronouns
def create_pronoun_indicators(x, pronoun_set):
    x_lower = x.lower()  # Convert to lowercase for case-insensitive comparison
    return int(any(re.search(r'\b' + re.escape(pronoun) + r'\b', x_lower) for pronoun in feature_wordsets[pronoun_set]))

# Apply the function to create indicator variables for each set of pronouns
for pronoun_set in feature_wordsets.keys():
    pair_df[f'{pronoun_set}1'] = pair_df['headline1'].apply(lambda x: create_pronoun_indicators(x, pronoun_set))
    pair_df[f'{pronoun_set}2'] = pair_df['headline2'].apply(lambda x: create_pronoun_indicators(x, pronoun_set))

# Display the updated dataframe with the new indicator variables
print(pair_df[['eyecatcher_id', 'lede', 'excerpt', 'headline1', 'numwords1', 'headline2', 'numwords2',
               'ctr1', 'ctr2', 'date_created',
               'first_person_singular1', 'first_person_singular2',
               'first_person_plural1', 'first_person_plural2',
               'second_person1', 'second_person2',
               'third_person_singular1', 'third_person_singular2',
               'third_person_plural1', 'third_person_plural2']])
```

## 3 Estimating the Effect of Language on Headline Success

3.1 First, we examine whether the winner headlines have more or fewer words than the loser headline. Conduct an independent sample t-test and paired t-test (see [scipy.stats](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html#scipy.stats.ttest_ind); for the independent sample t-test, assume equal variance). Also, calculate and print the mean difference between the number of words in the winner and the loser headlines.

```python
from scipy.stats import ttest_ind, ttest_rel

# Independent sample t-test (assuming equal variance)
ind_ttest_result = ttest_ind(pair_df['numwords1'], pair_df['numwords2'], equal_var=True)

# Paired t-test
paired_ttest_result = ttest_rel(pair_df['numwords1'], pair_df['numwords2'])

# Calculate and print the mean difference
mean_difference = pair_df['numwords1'].mean() - pair_df['numwords2'].mean()
print(f"Mean Difference: {mean_difference}")

# Print the t-test results
print("Independent Sample t-test:")
print(f"T-statistic: {ind_ttest_result.statistic}")
print(f"P-value: {ind_ttest_result.pvalue}")

print("\nPaired t-test:")
print(f"T-statistic: {paired_ttest_result.statistic}")
print(f"P-value: {paired_ttest_result.pvalue}")
```

3.7 **Discuss:** When are paired t-tests helpful? Justify.

**Paired t-tests** are helpful in situations where you have pairs of related observations. The key characteristic of paired data is that each observation in one group is paired with a unique observation in another group. Paired t-tests are specifically designed for comparing means of two related groups, taking into account the dependency between observations. Here are some scenarios where paired t-tests are particularly useful:

1. **Repeated Measures Design:**
   - Paired t-tests are commonly used in experimental designs where the same subjects are used for each treatment or condition.
   - For example, measuring the performance of individuals before and after a treatment, intervention, or exposure.

2. **Matched Samples:**
   - When subjects are matched based on certain characteristics, and measurements are taken before and after an intervention, a paired t-test is appropriate.
   - For instance, comparing the blood pressure of individuals before and after a specific medication, where each individual serves as their own control.

3. **Natural Pairs:**
   - In cases where natural pairs exist in the data, and it makes sense to compare observations within each pair.
   - This could be siblings, twins, or matched pairs based on certain criteria.

4. **Reducing Variability:**
   - Paired t-tests are often more powerful than independent sample t-tests because they reduce variability by accounting for the pairing.
   - This increased power can be particularly beneficial in situations with small sample sizes.

So, paired t-tests are helpful when dealing with related observations and the goal is to compare the mean differences within pairs. They provide a more sensitive analysis in situations where individual differences or extraneous variables might affect the outcome, making them a valuable tool in various fields, including medicine, psychology, and experimental sciences.

Here, note that we are considering "at least one positive outcome" to be the manifestation of an underlying effect, thus significance level must be adjusted down when performing multiple hypothesis tests!

```python
from scipy.stats import ttest_ind

# Assuming you have separate DataFrames for t1 and t2: t1_df, t2_df

# Define the list of effects
effects = ['num_words', 'first_person_singular', 'first_person_plural', 'second_person', 'third_person_singular', 'third_person_plural']

# Add positive and negative effects separately
positive_effects = [f'positive{i}' for i in range(1, 3)]
negative_effects = [f'negative{i}' for i in range(1, 3)]
effects += positive_effects + negative_effects

# Initialize a dictionary to store results
results = {}

# Perform hypothesis tests for each effect
for effect in effects:
    # Extract data for the effect from t1 and t2
    effect_t1 = t1_df.filter(like=effect).sum(axis=1)
    effect_t2 = t2_df.filter(like=effect).sum(axis=1)

    # Perform t-test
    t_statistic, p_value = ttest_ind(effect_t1, effect_t2)

    # Store results in the dictionary
    results[effect] = {'t_statistic': t_statistic, 'p_value': p_value}

# Adjust significance level for multiple testing (Bonferroni correction)
num_tests = len(effects)
adjusted_significance_level = 0.05 / num_tests

# Evaluate significance for each effect
for effect, result in results.items():
    if result['p_value'] < adjusted_significance_level:
        print(f"The difference in {effect} between t1 and t2 is statistically significant.")
    else:
        print(f"There is no significant difference in {effect} between t1 and t2.")
```
